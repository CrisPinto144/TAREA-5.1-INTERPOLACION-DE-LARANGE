```python
#   Codigo que implementa la interpolacion de Lagrange
#   para ajustar un conjunto de datos
#
#
#           Autor:
#   Cristian Yuseff Pinto Villanueva
#   cristian.pino984@gmail.com
#   Version 1.0 : 01/04/2025
#
#   EJERCICIO 1

import numpy as np
import matplotlib.pyplot as plt

# Datos del ejercicio: posici√≥n (m) y deformaci√≥n (mm)
x_points = np.array([0.5, 1.0, 1.5, 2.0])
y_points = np.array([1.2, 2.3, 3.7, 5.2])

# Funci√≥n de interpolaci√≥n de Lagrange
def lagrange_interpolation(x, x_points, y_points):
    n = len(x_points)
    result = 0
    for i in range(n):
        term = y_points[i]
        for j in range(n):
            if i != j:
                term *= (x - x_points[j]) / (x_points[i] - x_points[j])
        result += term
    return result

# a) Calcular la deformaci√≥n en x = 1.25 m
x_eval = 1.25
deformacion_estimada = lagrange_interpolation(x_eval, x_points, y_points)
print(f"Deformaci√≥n estimada en x = {x_eval} m: {deformacion_estimada:.4f} mm")

# b) Graficar la interpolaci√≥n en el rango [0.5, 2.0]
x_values = np.linspace(0.5, 2.0, 200)
y_values = [lagrange_interpolation(x, x_points, y_points) for x in x_values]

# Graficar
plt.figure(figsize=(6,4))
plt.plot(x_values, y_values, label="Interpolaci√≥n de Lagrange", color="blue")
plt.scatter(x_points, y_points, color="red", label="Puntos dados")
plt.scatter(x_eval, deformacion_estimada, color="green", label=f"Estimaci√≥n en x={x_eval}", zorder=5)
plt.xlabel("Posici√≥n (m)")
plt.ylabel("Deformaci√≥n (mm)")
plt.title("Interpolaci√≥n de Lagrange - Deformaci√≥n en una Viga")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
